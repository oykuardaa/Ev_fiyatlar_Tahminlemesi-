```python
#Underfitting: Underfitting, bir modelin eğitim verilerine yeterince uyum sağlayamadığı durumu ifade eder. Bu durumda model, hem eğitim hem de test verileri üzerinde düşük performans gösterebilir. Model çok basit olduğunda veya veri seti karmaşıklığına uygun değilse underfitting görülebilir.
#Overfitting: Overfitting, bir modelin eğitim verilerine aşırı uyum sağladığı ve genelleştirme yapmakta zorlandığı durumu ifade eder. Model eğitim verilerini ezberler gibi öğrenir ve yeni veriler üzerinde kötü sonuçlar üretebilir. Modelin karmaşıklığı yüksek olduğunda veya veri seti çok küçük olduğunda overfitting görülebilir.
##Veri ön işleme aşamasında kategorik sütunları sayısal değerlere dönüştürmeniz gerekebilir. Bunun için sıklıkla kullanılan bir yöntem "One-Hot Encoding"dir. Bu işlem, kategorik sütunları binary sütunlara (0 ve 1'lerden oluşan) dönüştürür
#
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
# Veri setini yükledik
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
# Eksik değerleri doldurmak istediğimiz sütünlar
columns_to_fill = ['LotFrontage', 'MasVnrArea', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                   'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType', 'GarageYrBlt', 'GarageFinish', 
                   'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']

# Kategorik sütunlardaki eksik değerleri uygun değerle doldurduk 
for col in columns_to_fill:
    if data[col].dtype == 'object':
        most_common_value = data[col].mode()[0]
        data[col].fillna(most_common_value, inplace=True)

# Sayısal sütunlardaki eksik değerleri medyan ile doldurduk
for col in columns_to_fill:
    if data[col].dtype != 'object':
        median_value = data[col].median()
        data[col].fillna(median_value, inplace=True)

data
```

```python
# Genel durum puanlamasının fiyatla ilişkisini inceleyen bir grafik çizdik
plt.figure(figsize=(10, 6))
sns.boxplot(x="OverallCond", y="SalePrice", data=data)
plt.xlabel("Genel Durum")
plt.ylabel("Satış Fiyatı")
plt.title("Genel Durum ve Satış Fiyatı İlişkisi")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bağımsız değişken (X) olarak OverallCond'u seçtik
X = data[["OverallCond"]]

# Bağımlı değişken (y) olarak SalePrice'ı seçtik
y = data["SalePrice"]

# Modeli oluştur ve eğit
model = LinearRegression()
model.fit(X, y)

# Eğitilmiş model ile tahmin yap
y_pred = model.predict(X)

# Model performansını değerlendir
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Sonuçları yazdır
print("Mean Squared Error:", mse)
print("R2 Score:", r2)

# Veriyi görselleştir
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red', linewidth=3)
plt.xlabel("Overall Condition")
plt.ylabel("Sale Price")
plt.title("Regression Analysis: Overall Condition vs. Sale Price")
plt.show()
#Çok anlamlı bir sonuç elde edemedik o yüzden başka bağımsız girdiler ile analiz yappmak daha mantıklı olacaktır.
```

```python
#oda sayısı ve fiyat arasındaki ilişki grafiği
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotRmsAbvGrd', y='SalePrice', data=data)
plt.title('Oda Sayısı vs. Fiyat Scatter Plot')
plt.xlabel('Oda Sayısı')
plt.ylabel('Fiyat')
plt.show()
```

```python
# Bağımsız değişken olarak oda sayısı (TotRmsAbvGrd) ve bağımlı değişken olarak fiyatı (SalePrice) seçtik
X1 = data[['TotRmsAbvGrd']]
y1= data['SalePrice']
```

```python
# Veriyi eğitim ve test setlere ayır
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
# Linear Regression modelini oluşturduk
model = LinearRegression()

# Modeli eğittik
model.fit(X1_train, y1_train)

# Test verilerini kullanarak tahmin yaptık
y_pred = model.predict(X1_test)

# Model performansını değerlendirdik
mse = mean_squared_error(y1_test, y_pred)
r2 = r2_score(y1_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# Sonuçları bir arada görselleştirdik
plt.figure(figsize=(10, 6))
plt.scatter(X1_test, y1_test, color='blue', label='Gerçek Değerler')
plt.plot(X1_test, y_pred, color='red', linewidth=2, label='Tahmin Edilen Değerler')
plt.title('Linear Regression: Oda Sayısı vs. Fiyat')
plt.xlabel('Oda Sayısı')
plt.ylabel('Fiyat')
plt.legend()
plt.show()
```

```python
# Bağımsız değişkenler (X) ve bağımlı değişken (y) olarak ayırdık burda ev özelliklerine göre fiyatların tahminlemesini yapmaya çalışacağız
y2 = data["SalePrice"]
X2 = data.drop("SalePrice", axis=1)

# Kategorik sütunları One-Hot Encoding ile dönüştürmeliyiz
X2_encoded = pd.get_dummies(X2, drop_first=True)

# Eksik verileri doldurduk 
X2_encoded.fillna(0, inplace=True)  # Eksik verileri 0 ile doldurduk, daha iyi yöntemler de kullanabilirsiniz

# Veriyi eğitim ve test setlerine ayırdık
X2_train, X2_test, y2_train, y2_test = train_test_split(X2_encoded, y2, test_size=0.2, random_state=42)

# Modeli oluştur ve eğittik
model = RandomForestRegressor(random_state=42)
model.fit(X2_train, y2_train)

# Test verileri üzerinde tahmin yaptık
y_pred2 = model.predict(X2_test)

# Model performansını değerlendirdik
mse = mean_squared_error(y2_test, y_pred2)
r2_2 = r2_score(y2_test, y_pred2)

print("Mean Squared Error:", mse)
print("R2 Score:", r2_2)
```
